```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df_new = df.drop("target", axis = 1)
```

```python
test = pd.read_csv("test.csv")
```

```python
test.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
target = df["target"]
```

```python
df.isnull().sum(axis=0)
```

```python
df3 = df_new
```

```python
df3 = df3.drop("DepTime", axis = 1)
df3 = df3.drop("AirTime", axis = 1)
df3 = df3.drop("ArrTime", axis = 1)
df3 = df3.drop("TailNum", axis = 1)
df3 = df3.drop("TaxiIn", axis = 1)
df3 = df3.drop("TaxiOut", axis = 1)
df3 = df3.drop("Id", axis = 1)
df3 = df3.drop("Dest", axis = 1)
df3 = df3.drop("Origin", axis = 1)
df3 = df3.drop("UniqueCarrier", axis = 1)
df3.fillna(0)
```

```python
logr = logreg.fit(df3, target)
```

```python
test = test.drop("DepTime", axis = 1)
test = test.drop("ArrTime", axis = 1)
test = test.drop("AirTime", axis = 1)
test = test.drop("TaxiIn", axis = 1)
test = test.drop("TaxiOut", axis = 1)
test = test.drop("UniqueCarrier", axis = 1)
test = test.drop("TailNum", axis = 1)
test = test.drop("Origin", axis = 1)
test = test.drop("Dest", axis = 1)
test = test.drop("Id", axis = 1)
```

```python
test = test.fillna(0)
```

```python
test.head()
```

```python
proba = logr.predict_proba(test)[:, 1]
```

```python
submission = pd.read_csv("sample_submission.csv")
```

```python
submission['target'] = proba
```

```python
submission.to_csv("submission", index = False)
```
